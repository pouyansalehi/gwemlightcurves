# <span style="color:green"> 1) Simulating a Kilonova Lightcurve</span> 

When attempting to simulate the lightcurve from a kilonova assosciated with a binary coalesence, a number of parameters must be determine first:
- Firstly, determine if the system is a BNS or a NSBH system (which fit model should be used)
- Fit model will compute 
  > the expected dynamical ejecta mass <br>
  > dynamical ejecta velocity .. that arises from the system.
  
For binary neutron stars, one can use Tim Dietrich and Maximiliano Ujevic, and for a NSBH one can use Kawaguchi. 

These fits tie information such as:
- compactness, 
- baryonic mass, 
- mass of the objects to the expected ejecta

These fits come with uncertainities of ~72% but **knowing the ejecta mass is critical when determining the expected lightcurve from the resulting kilonova.** 

We now explain how one can go from a set of posteriors of the system to generating the expected lightcurve from the kilonova. We consider many important things including:
- whether or not you make Equation of State (EOS) assumptions
- whether or not you use fits for compactness and baryonic mass, and
- how you treat the uncertainity in ejecta that comes with the fit.

#### 1.4) Plotting the Ejecta Mass and Velocity of Ejecta

Now that we have compactness and the baryonic mass we can calculate **Mass of the ejecta** and the **velocity of the ejecta** using fits from Tim Dietrich and Maximiliano Ujevic.

The **dynamical ejecta mass fit formula** can be found

https://arxiv.org/pdf/1612.03665.pdf#equation.3.1

\begin{equation*}
\frac{M_{ej}^{fit}}{10^{-3} M_{\odot}} = \left[a \Big( \frac{M_2}{M_1} \Big)^{1/3}  \Big( \frac{1 - 2C_1}{C_1} \Big) + b\Big( \frac{M_2}{M_1} \Big)^{n} + c \Big( 1- \frac{M_1}{M_1^*} \Big) \right] M_1^* + (1 \leftrightarrow 2) + d
\label{eq:vector_ray} \tag{1}
\end{equation*}

and the constants are taken from

https://arxiv.org/pdf/1612.03665.pdf#equation.3.2

```python
a = −1.35695 
b = 6.11252
c = -49.43355 
d = 16.1144 
n = −2.5484
```

The method used to calculate in this repo is `gwemlightcurves.EjectaFits.DiUj2017.calc_meje()` and can be used as follows:

```python
from gwemlightcurves.EjectaFits.DiUj2017 import calc_meje
from gwemlightcurves.KNModels import KNTable

t = KNTable.read_samples('posterior_samples.dat')
t = t.calc_tidal_lambda(remove_negative_lambda=True)
t_sly_mon = t.calc_radius(EOS='sly', TOV='Monica')
t_sly_mon = t_sly_mon.calc_compactness()
t_sly_mon = t_sly_mon.calc_baryonic_mass(EOS='sly', TOV='Monica')

# here use new function calc_meje()
t_sly_mon['mej'] = calc_meje(t_sly_mon['m1'], t_sly_mon['mb1'], t_sly_mon['c1'], t_sly_mon['m2'], t_sly_mon['mb2'], t_sly_mon['c2'])
```

```python
# calculated values of Ejecta Mass in Msol already divided by 10^{-3}
t_sly_mon['mej']
```

The **velocity of the ejecta mass** fit can be found:

https://arxiv.org/pdf/1612.03665.pdf#equation.3.9

\begin{equation*}
v_{ej} = \sqrt{v_{\rho}^{2} + v_z^2}
\label{eq:vector_ray} \tag{2}
\end{equation*}

where $v_{\rho}$ is an EOS dependent fitting function by including a first order polynomial depending on the
compactness $(1 + c C_{1,2})$, which leads to

\begin{equation*}
v_{\rho} =  \left[ a \Big( \frac{M_1}{M_2} \Big) (1 + c C_1) \right] + (1 \leftrightarrow 2) + b 
\label{eq:vector_ray} \tag{2}
\end{equation*}

where parameters are: $a = −0.219479, b = 0.444836$ and $ c = −2.67385$.

and $v_z$ is the velocity orthogonal to the orbital plane and is given as:

\begin{equation*}
v_{z} =  \left[ a \Big( \frac{M_1}{M_2} \Big) (1 + c C_1) \right] + (1 \leftrightarrow 2) + b 
\label{eq:vector_ray} \tag{2}
\end{equation*}

where parameters are $a = −0.315585$, $b = 0.63808$ and $c = −1.00757$.

The method used to calculate in this repo is `gwemlightcurves.EjectaFits.DiUj2017.calc_vej()` and can be used as follows:

```python
from gwemlightcurves.EjectaFits.DiUj2017 import calc_vej
t_sly_mon['mej'] = calc_vej(t_sly_mon['m1'], t_sly_mon['c1'], t_sly_mon['m2'], t_sly_mon['c2'])
```

#### 1.5) Generating Lightcurves

Finally, let’s **calculate a lightcurve** being **EOS agnostic**, meaning we calculate both the compactness and baryonic masses from fits. 

Also let us look at:
- Metzer 2017, and 
- DiUj2017<br>

models. 

In order to take a set of samples and calculate the light curves that would result from a realization of each sample, you can use the model which takes as inputs:
- the string name of the model, and 
- the table of samples containing at minimum compactness and baryonic mass (it can calculate mass ejecta and velocity of ejecta on the fly).

```python
from gwemlightcurves.KNModels import KNTable
from gwemlightcurves import lightcurve_utils

t = KNTable.read_samples('posterior_samples.dat')
t = t.calc_tidal_lambda(remove_negative_lambda=True)
t = t.calc_compactness(fit=True)
t = t.calc_baryonic_mass(EOS=None, TOV=None, fit=True)
t = t.downsample(Nsamples=100)

tini = 0.1;  # initial time
tmax = 50.0; # max. time
dt = 0.1;    # time step
vmin = 0.02; # min. velocity
th = 0.2;    # ??
ph = 3.14;   # ??
kappa = 1.0; # opacity
eps = 1.58*(10**10); # ??
alp = 1.2;   # ??
eth = 0.5;   # ?? thermal energy?
flgbct = 1;  # ??
beta = 3.0;  # ??
kappa_r = 1.0; # ??
slope_r = -1.2; # slope for what
theta_r = 0.0; # ??
Ye = 0.3 # electron fraction

# provides these values to keys
t['tini'] = tini; t['tmax'] = tmax; t['dt'] = dt; t['vmin'] = vmin; t['th'] = th; t['ph'] = ph; t['kappa'] = kappa; t['eps'] = eps; t['alp'] = alp; t['eth'] = eth; t['flgbct'] = flgbct; t['beta'] = beta; t['kappa_r'] = kappa_r; t['slope_r'] = slope_r; t['theta_r'] = theta_r; t['Ye'] = Ye
```

```python
# Create dict of tables for the various models, calculating mass ejecta velocity of ejecta and the lightcurve from the model
models = ["DiUj2017","Me2017"]
model_tables = {}
for model in models:
    model_tables[model] = KNTable.model(model, t)

# Now we need to do some interpolation
for model in models:
    model_tables[model] = lightcurve_utils.calc_peak_mags(model_tables[model])
    model_tables[model] = lightcurve_utils.interpolate_mags_lbol(model_tables[model])
```

#### 3.4) Direct Comparison

Here we plot a direct comparison of mass radius curves from the three different TOV solvers.



```python
from distutils.spawn import find_executable
from gwemlightcurves.KNModels import table
from gwpy.table import EventTable
#from gwpy.plotter import EventTablePlot
import astropy.units as u
import astropy.constants as C


G = C.G.value; c = C.c.value; msun = u.M_sun.to(u.kg)
```

```python
#plot = EventTablePlot(figsize=(20.5, 10.5))
EOS = ['ap4', 'H4', 'ms1b']
Color = ['blue', 'green', 'red']
locations = [(1,3,1), (1,3,2), (1,3,3)]
plot_location = dict(zip(EOS, locations))
```

```python
for eos in EOS:
    ax = plot.add_subplot(plot_location[eos][0], plot_location[eos][1], plot_location[eos][2])
    ax.set_title('EOS: {0}'.format(eos), fontsize='small')
    t_mon=EventTable.read(find_executable(eos+'_mr.dat'), format='ascii')
    t_wk=EventTable.read(find_executable(eos+'.tidal.seq'), format='ascii')
    t_lalsim=EventTable.read(find_executable(eos+'_lalsim_mr.dat'), format='ascii')
    wk_conversion=(msun * G / c**2)*10**-3
    mask_mon=t_mon['radius']<20
    mask_wk=t_wk['Circumferential_radius']<20
    mask_lalsim=t_lalsim['radius']<20
    plot.add_scatter(t_mon['radius'][mask_mon], t_mon['mass'][mask_mon], label='Monica '+eos ,color=Color[0], ax=ax)
    plot.add_scatter(t_wk['Circumferential_radius'][mask_wk]*wk_conversion, t_wk['grav_mass'][mask_wk], label='Wolfgang '+eos ,color=Color[1], ax=ax)
    plot.add_scatter(t_lalsim['radius'][mask_lalsim], t_lalsim['mass'][mask_lalsim], label='lalsim '+eos ,color=Color[2], ax=ax)
    plot.add_legend(loc="upper left", fancybox=True, fontsize='small')
```

```python
plot.text(0.5, 0.04, 'Radius (km)', ha='center', fontsize='x-large')
plot.text(0.04, 0.5, 'Mass ($M_{\odot}$)', va='center', rotation='vertical', fontsize='x-large')
plot.suptitle('Mass-Radius Curve Comparison', fontsize='x-large')
```
